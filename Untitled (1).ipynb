```python
boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard','Mick'] 
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']
if len(boys)==len(girls):
    sorted_boys = sorted(boys)
    sorted_girls = sorted(girls)
    for i in range(len(sorted_boys)):
        print(sorted_boys[i], ' and ',sorted_girls[i] )
else:
    print('мы не будем никого знакомить, если кто-то может остаться без пары:)')
```
